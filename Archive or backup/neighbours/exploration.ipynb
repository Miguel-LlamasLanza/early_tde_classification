```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tde_analysis as td
```

```python
data, features = td.get_data('features_all.csv')
data.loc[data['type']!='TDE','type'] = 'other'
```

```python
ft = 'temperature'

plt.figure()
plt.title(ft)

zoomed = (data[ft]>0) & (data[ft]<25000)
sns.kdeplot(data[zoomed], x=ft,  hue='type', common_norm=False)
```

```python
ft = 'amplitude'
plt.figure()
plt.title(ft)

zoomed = (data[ft]>0) & (data[ft]<100)
sns.kdeplot(data[zoomed], x=ft,  hue='type', common_norm=False)
```
